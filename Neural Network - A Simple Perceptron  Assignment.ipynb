# Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.
# Answer: 

### 1. Definition of Deep Learning
- **Deep Learning (DL)** is a subset of Machine Learning (ML) that uses **artificial neural networks with many layers** (deep architectures) to automatically learn hierarchical features from data.
- It excels in handling large, unstructured data such as **images, audio, text, and video**.

---

### 2. Evolution of Deep Learning
- **1950s–1980s:** Basic perceptron models and early neural networks were proposed, but limited by computational power and small datasets.  
- **1990s:** Neural networks lost popularity due to issues like **vanishing gradients** and preference for simpler ML algorithms (SVM, Decision Trees).  
- **2006 onwards:** Geoffrey Hinton and others introduced **deep neural networks** with improved training methods (e.g., backpropagation, ReLU).  
- **2010s:** Explosion of **big data**, **GPUs**, and **large labeled datasets** (like ImageNet) made DL practical and highly successful.  
- **Today:** Deep Learning powers applications like **speech recognition, computer vision, NLP, autonomous vehicles, and generative AI**.

---

### 3. Difference between Deep Learning and Traditional Machine Learning

| Aspect               | Traditional Machine Learning                  | Deep Learning                                |
|----------------------|-----------------------------------------------|----------------------------------------------|
| Feature Engineering  | Manual – humans select features from data     | Automatic – model learns features itself      |
| Data Requirement     | Works well with small/medium datasets         | Requires very large datasets                 |
| Performance          | Limited on complex/unstructured data          | Excellent on images, text, audio, video      |
| Computation          | Less computationally expensive                | Needs high computation (GPUs/TPUs)           |
| Interpretability     | Easier to interpret (decision trees, regression) | Harder to interpret (black-box models)    |

---

### 4. Summary
- **Deep Learning** evolved as a more powerful approach than traditional ML by leveraging **multi-layer neural networks**, **large datasets**, and **advanced hardware**.
- The key difference is **automatic feature extraction**:  
  - Traditional ML relies on human-designed features.  
  - Deep Learning directly learns complex patterns from raw data.

# Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.
**Answer.**  
Activation functions introduce **non-linearity** into neural networks. Without them, the network would behave like a linear model and fail to learn complex patterns. They decide whether a neuron should be activated based on weighted inputs and bias.  

- **Purpose of Activation Function:**  
  - Introduce **non-linearity** to handle complex data.  
  - Allow networks to learn **non-linear decision boundaries**.  
  - Control the **flow of information** and gradients during training.  

- **Comparison of Common Functions:**  

1. **Sigmoid Function:**  
   - Range: (0, 1)  
   - Smooth curve, good for probabilities.  
   - Limitation: Causes **vanishing gradient problem** for large values.  

2. **Tanh Function:**  
   - Range: (-1, 1)  
   - Centered at zero → better than sigmoid for optimization.  
   - Still suffers from **vanishing gradients** for extreme values.  

3. **ReLU (Rectified Linear Unit):**  
   - Range: [0, ∞)  
   - Very efficient, widely used in deep networks.  
   - Avoids vanishing gradient for positive values.  
   - Limitation: **Dying ReLU problem** (neurons stuck at zero).  

**In short,** Sigmoid and Tanh are useful for small/simple networks, while **ReLU is the most popular** in deep learning due to faster training and better performance.    

# Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.
# Answer:

In neural networks, both **loss function** and **cost function** are used to measure the performance of a model, but they are applied at different levels.

### 1. Loss Function
- A **loss function** quantifies the error for a **single training example**.
- It compares the predicted output with the actual (target) value.
- The goal of the loss function is to measure **how wrong** the model is on that one example.
- During training, the network computes the loss for each sample to update weights through **backpropagation**.

**Examples of Loss Functions:**
- **Mean Squared Error (MSE):** Common in regression tasks.  
  \( L = (y_{true} - y_{pred})^2 \)
- **Binary Cross-Entropy Loss:** Used in binary classification.  
  \( L = -[y \log(\hat{y}) + (1-y) \log(1-\hat{y})] \)

---

### 2. Cost Function
- A **cost function** is the **average of the loss values** over the entire training dataset (or mini-batch).
- It gives an overall measure of how well the model is performing across all samples.
- The cost function is minimized during training using optimization algorithms like **Gradient Descent**.

**Examples of Cost Functions:**
- **Average MSE:**  
  \( J(\theta) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 \)
- **Average Cross-Entropy:** Used in classification tasks.

---

### 3. Key Differences

| Aspect              | Loss Function                        | Cost Function                        |
|---------------------|--------------------------------------|---------------------------------------|
| Level               | Single training example              | Entire training dataset (average loss) |
| Purpose             | Measures individual prediction error | Measures overall model performance     |
| Use in Training     | Computed per sample                  | Guides weight updates globally         |
| Example             | Error of one student’s marks         | Average error of the whole class       |

---

### 4. Analogy
- Imagine checking exam papers:
  - **Loss function** = error in **one student’s paper**.  
  - **Cost function** = **average error of the whole class**.

---

### Final Note
- In practice, the **loss function** is computed for each data point, then aggregated to form the **cost function**, which is minimized during model training.
- Thus, *loss* is local (per sample), while *cost* is global (entire dataset).

# Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.
# Answer:

### 1. Role of Optimizers
- In neural networks, **optimizers** are algorithms used to **update the weights and biases** to minimize the **loss/cost function**.  
- They determine **how quickly and accurately** a model converges during training.  
- Without optimizers, the network cannot learn from data effectively.

---

### 2. Common Optimizers

| Optimizer           | How it Works                                                                 | Advantages                                      | Disadvantages                                |
|--------------------|----------------------------------------------------------------------------|------------------------------------------------|---------------------------------------------|
| **Gradient Descent (GD)** | Computes gradients of the loss w.r.t all weights and updates them in the negative gradient direction. | Simple and effective for small datasets       | Slow convergence, can get stuck in local minima |
| **RMSprop**         | Uses a moving average of squared gradients to **adapt learning rates** for each parameter. | Works well for non-stationary and online data | Slightly more complex than GD               |
| **Adam**            | Combines **momentum** (moving average of gradients) and RMSprop (adaptive learning rate). | Fast convergence, widely used, robust         | Can overfit if learning rate not tuned     |

---

### 3. Summary
- Optimizers are **critical for training neural networks efficiently**.  
- **Gradient Descent** → basic, simple.  
- **RMSprop** → adapts learning rate per parameter.  
- **Adam** → combines momentum + adaptive learning rate → often the best default choice.

# Question 6: Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.
# Answer:
    

```python
# Single-Layer Perceptron to solve AND gate using NumPy

import numpy as np

# Activation function (Step function)
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Input dataset for AND gate
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

# Target outputs
y = np.array([0, 0, 0, 1])

# Initialize weights and bias
weights = np.random.rand(2)
bias = np.random.rand(1)

# Learning rate
lr = 0.1

# Training the perceptron
for epoch in range(10):  # 10 iterations
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        y_pred = step_function(linear_output)
        # Update weights and bias
        weights += lr * (y[i] - y_pred) * X[i]
        bias += lr * (y[i] - y_pred)
    print(f"Epoch {epoch+1}: Weights={weights}, Bias={bias}")

# Testing
print("\nTesting AND gate:")
for i in range(len(X)):
    output = step_function(np.dot(X[i], weights) + bias)
    print(f"Input: {X[i]}, Predicted Output: {output}")
```

# Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.
# Answer:

```python
# Visualizing Sigmoid, ReLU, and Tanh activation functions

import numpy as np
import matplotlib.pyplot as plt

# Input range
x = np.linspace(-10, 10, 200)

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plotting
plt.figure(figsize=(12, 4))

# Sigmoid
plt.subplot(1, 3, 1)
plt.plot(x, y_sigmoid, color='blue')
plt.title("Sigmoid Function")
plt.grid(True)

# ReLU
plt.subplot(1, 3, 2)
plt.plot(x, y_relu, color='green')
plt.title("ReLU Function")
plt.grid(True)

# Tanh
plt.subplot(1, 3, 3)
plt.plot(x, y_tanh, color='red')
plt.title("Tanh Function")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Question 8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.
# Answer:

```python
import os
import warnings

# Suppress warnings
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
warnings.filterwarnings("ignore", category=UserWarning)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Input
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train/255.0, x_test/255.0
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build model
model = Sequential([
    Input(shape=(28,28)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train model
history = model.fit(x_train, y_train, epochs=5, batch_size=32, validation_split=0.1, verbose=1)

# Evaluate
test_loss, test_acc = model.evaluate(x_test, y_test)
print(f"\nTest Accuracy: {test_acc:.4f}")

```

# Question 9: Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.
# Answer: 

```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Normalize input data
x_train = x_train / 255.0
x_test = x_test / 255.0

# One-hot encode labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build a simple neural network
model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=32,
                    validation_split=0.1)

# Plot training & validation loss
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Training Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='o')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Plot training & validation accuracy
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Interpretation:
# 1. Training loss decreases and training accuracy increases steadily, indicating the model is learning.
# 2. Validation loss and accuracy follow a similar trend. 
# 3. If validation loss starts increasing while training loss decreases, it indicates overfitting.
# 4. This visualization helps identify model convergence, overfitting, or underfitting.
```

# Question 10: 
You are working on a project for a bank that wants to automatically detect fraudulent transactions. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate.
    
Explain your real-time data science workflow:
    
● How would you design a deep learning model (perceptron or multilayer NN)?

● Which activation function and loss function would you use, and why?

● How would you train and evaluate the model, considering class imbalance?

● Which optimizer would be suitable, and how would you prevent overfitting?

# Answer:



**1. Data Understanding & Preprocessing:**
- Analyze the dataset for feature distributions, missing values, and anomalies.
- Handle missing data: impute numeric features with mean/median, categorical with mode.
- Feature encoding: one-hot encode or embed categorical features like merchant ID and customer location.
- Feature scaling: normalize numeric features (e.g., transaction amount) using MinMaxScaler or StandardScaler.
- Train-test split: use stratified splitting to maintain class distribution.

---

**2. Model Design:**
- Use a **multilayer neural network (MLP)** because tabular transaction data often has nonlinear relationships.
    - Input layer: number of neurons = number of features.
    - Hidden layers: 2-3 layers with 64-128 neurons each.
    - Output layer: single neuron with **sigmoid activation** for binary classification.
- Activation functions:  
    - **ReLU** for hidden layers to enable non-linear learning and avoid vanishing gradients.  
    - **Sigmoid** for output to predict probability of fraud (0-1).

---

**3. Loss Function & Evaluation Metrics:**
- **Binary Cross-Entropy** loss for probability-based binary classification.
- Metrics: accuracy alone is insufficient for imbalanced data; use:
    - Precision, Recall, F1-score  
    - ROC-AUC  
    - Confusion matrix to evaluate false positives and false negatives.

---

**4. Handling Class Imbalance:**
- **Resampling techniques:**  
    - Oversample minority class using SMOTE.  
    - Undersample majority class if needed.  
- **Class weighting:** assign higher weight to minority class in loss function to penalize misclassification of fraudulent transactions.

---

**5. Optimizer:**
- **Adam optimizer** for adaptive learning rates and fast convergence.

---

**6. Preventing Overfitting:**
- L2 regularization to penalize large weights.
- Dropout layers (0.2-0.5) between hidden layers.
- Early stopping based on validation loss.
- Batch normalization to improve training stability and generalization.

---

**7. Training & Evaluation Workflow:**
1. Preprocess and balance dataset.
2. Train MLP using class weighting and validation split.
3. Monitor training and validation loss and metrics.
4. Apply early stopping to prevent overfitting.
5. Evaluate the final model on test data using precision, recall, F1-score, and ROC-AUC.
